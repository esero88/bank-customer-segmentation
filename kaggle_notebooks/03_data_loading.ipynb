```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
df = pd.read_parquet("../data/processed/clean_transactions.parquet")
```

```python
df.head()
```

```python
df = df.rename(columns={
    "TransactionID": "transaction_id",
    "CustomerID": "customer_id",
    "CustomerDOB": "customer_dob",
    "CustGender": "customer_gender",
    "CustLocation": "customer_location",
    "CustAccountBalance": "customer_account_balance",
    "TransactionDate": "transaction_date",
    "TransactionTime": "transaction_time",
    "TransactionAmount (INR)": "transaction_amount_inr"
})
```

```python
engine = create_engine("postgresql://postgres:1234@localhost:5432/bank_dwh")
```

```python
df.to_sql(
    "transactions_clean",
    engine,
    schema="staging",
    if_exists="append",
    index=False
)
```
